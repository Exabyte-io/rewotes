```python
import pandas as pd
import json   
```

```python
MP_output_dir = "../data/training/materialsproject_output/"
MP_json_dir = "../data/training/materialsproject_json/"
fname = "Si"
```

```python
data_df = pd.read_csv(f"{MP_output_dir}/{fname}.csv")
data_df.head()
```

```python
data_df = data_df.drop('Unnamed: 13', axis=1)
data_df.columns = [
    "material_ID",
    "formula",
    "spacegroup",
    "formation_energy__eV",
    "E_above_hull__eV",
    "band_gap__eV",
    "has_bandstructure",
    "volume",
    "Nsites",
    "theoretical",
    "count",
    "density__gm_per_cc",
    "crystal_system",
]
data_df = data_df.set_index('material_ID')
data_df.head()
```

```python
data_dic = data_df.to_dict('index')
with open(f"{MP_json_dir}/{fname}.json","w") as file:
    json.dump(data_dic,file,indent=4)
```
