```python
# https://gist.github.com/blondegeek/f96d31642bc804406eef52195eedb61f?short_path=1afc78c
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.groups import SpaceGroup, SYMM_DATA, sg_symbol_from_int_number
```

```python
# To make Materials Project queries, you need to have your Materials Project API key set in ~/.pmgrc.yaml with:
# PMG_MAPI_KEY: YOUR_API_KEY
a = MPRester()
```

```python
import pymatgen as mg
```

```python
sg_int = [SpaceGroup('Pa-3').int_number]
properties = ["material_id","pretty_formula","band_gap","cif"]

results = a.query(criteria={'spacegroup.number': {'$in': sg_int}}, properties=properties)
print("Number of entries in Materials Project w/ spacegroup Pa-3:",len(results))
```
